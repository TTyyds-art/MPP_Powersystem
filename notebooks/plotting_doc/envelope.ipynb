```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x_L, x_U = 0, 3
y_L, y_U = 0, 2

x = np.linspace(x_L, x_U, 100)
y = np.linspace(y_L, y_U, 100)
X, Y = np.meshgrid(x, y)

Z = X * Y

Z1 = x_L * Y + y_L * X - x_L * y_L
Z2 = x_U * Y + y_U * X - x_U * y_U
Z3 = x_L * Y + y_U * X - x_L * y_U
Z4 = x_U * Y + y_L * X - x_U * y_L

fig, axs = plt.subplots(2, 2, figsize=(12, 12), subplot_kw={'projection': '3d'})

angles = [
    {'elev': 30, 'azim': -60},
    {'elev': 30, 'azim': 60},
    {'elev': 30, 'azim': 120},
    {'elev': 30, 'azim': -120}
]

for ax, angle in zip(axs.flat, angles):
    ax.plot_surface(X, Y, Z, alpha=0.3, cmap='viridis')

    # Plot the edges of the wireframes
    for i in range(0, 100, 99):
        ax.plot(X[i, :], Y[i, :], Z1[i, :], color='red')
        ax.plot(X[i, :], Y[i, :], Z2[i, :], color='blue')
        ax.plot(X[i, :], Y[i, :], Z3[i, :], color='gray')
        ax.plot(X[i, :], Y[i, :], Z4[i, :], color='green')
        ax.plot(X[:, i], Y[:, i], Z1[:, i], color='red')
        ax.plot(X[:, i], Y[:, i], Z2[:, i], color='blue')
        ax.plot(X[:, i], Y[:, i], Z3[:, i], color='gray')
        ax.plot(X[:, i], Y[:, i], Z4[:, i], color='green')

    ax.view_init(elev=angle['elev'], azim=angle['azim'])
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('z')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_L = 2
x_U = 6
x = np.linspace(x_L, x_U, 1000)

# Square function
square_x = x**2

# Lower bound (square of the variable x)
lower_bound = x**2

# Upper bound (x_U + x_L)x - x_U * x_L
upper_bound = (x_U + x_L) * x - x_U * x_L

plt.plot(x, square_x, label='x^2 (square function)')
plt.plot(x, lower_bound, label='Lower bound (x^2)')
plt.plot(x, upper_bound, label='Upper bound ((x_U + x_L)x - x_U * x_L)')

plt.xlabel('x')
plt.ylabel('W')
plt.legend()
plt.title('Lower and Upper Bounds for W')
plt.grid(True)
plt.show()
```
